#10) What are the maximum number of subscribers in each level of courses

```
data.sort_values('level', ascending=False)
```

```
data.groupby('level').num_subscribers.sum()
```

```
data.groupby('level').num_subscribers.sum().sort_values(ascending=False)
```

```
data.groupby('level').num_subscribers.max()
```

```
data.groupby('level').num_subscribers.min()
```

```
data.groupby('level').max()
```
